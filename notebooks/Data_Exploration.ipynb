```python
for j in range(len(arr_cpub)):
  for i in range(len(df_pubs['Publisher Name'])):
    if (str(arr_cpub[j]) == df_pubs['Publisher Name'][i]): #direct matches ~50
      arr_matches.append(i)
    elif ((str(arr_cpub[j])+" ") in df_pubs['Publisher Name'][i]): #potential matches that contain substring, ~600
      print(str(arr_cpub[j])+' AND '+df_pubs['Publisher Name'][i])
      arr_pmatches.append(i)
```
